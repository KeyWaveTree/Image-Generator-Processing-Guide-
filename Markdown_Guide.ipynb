# **Markdown 기본 문법**
---
## 제목(header)
*   \#      제목 1(가장 큰제목)
*   \##     제목 2
*   \###    제목 3
*   \####   제목 4
*   \#####  제목 5
*   \###### 제목 6

\# 과 제목 텍스트 사이에 반드시 공백을 넣어야한다.

---

## 강조 표현


*   \*기울임 (italic)\* 혹은 \_기울임(italic)\_
*   \** 굵게 (bold) \** 혹은 \_ _ 굵게 (bold)\_ _
*   \*** 굵고 기울임 \*** 혹은 \_ _ _ 굵고 기울임 \_ _ _
*   \~~ 취소선 \~~


---

## 순서 없는 목록

* \* 항목 1
* \* 항목 2

  *  tab->| \* 하위 항목 2-1
  *  tab->| \* 하위 항목 2-2

* \* 항목 3
* \+ 또는 플러스 사용
* \- 또는 하이픈 사용

---

## 순서 있는 목록
* 1. 첫번째 항목
* 2. 두번째 항목
*   tab->| 1. 하위 항목 2-1
*   tab->| 2. 하위 항목 2-2
* 3. 세 번째 항목
